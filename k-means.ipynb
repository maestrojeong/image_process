```python
import numpy as np
import matplotlib.pyplot as plt
from ops import rgb2gray
from ops import k_means
from ops import freq_filter
```

```python
image = plt.imread('./image.png')
```

```python
gray_image = rgb2gray(image)
```

```python
low_pass, high_pass = freq_filter(gray_image, 20.0)
```

```python
k_means_image = k_means(gray_image, clusters=5)
```

```python
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
ax1.imshow(gray_image, cmap = 'gray')
ax1.set_axis_off()
ax1.set_title("original image")
ax2.imshow(high_pass, cmap = 'gray')
ax2.set_axis_off()
ax2.set_title("High pass image")
ax3.imshow(k_means_image, cmap = 'gray')
ax3.set_axis_off()
ax3.set_title("K means image")
plt.show()
```
