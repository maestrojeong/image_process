```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread('./flower.png')
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
height, width, channel = image.shape
```

```python
R_image = np.zeros((height,width, channel))
G_image = np.zeros((height,width, channel))
B_image = np.zeros((height,width, channel))
for h in range(height):
    for w in range(width):
        R_image[h][w][0] = image[h][w][0]
        G_image[h][w][1] = image[h][w][1]
        B_image[h][w][2] = image[h][w][2]
```

```python
H = np.zeros((height, width))
S = np.zeros((height, width))
V = np.zeros((height, width))
for h in range(height):
    for w in range(width):
        R = image[h][w][0]
        G = image[h][w][1]
        B = image[h][w][2]
        M = np.max(image[h,w,:])
        m = np.min(image[h,w,:])
        D = M - m
        if D==0:
            H[h][w] = 0
        elif M == R:
             H[h][w] = 60*(G-B)/D
        elif M == G:
             H[h][w] = 60*(B-R)/D
        elif M == B:
             H[h][w] = 60*(R-G)/D
        if M==0:
            V[h][w] = 0
            S[h][w] = 0
        else:
            V[h][w] = M
            S[h][w] = D/V[h][w]
```

```python
fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)
ax4 = fig.add_subplot(2, 3, 4)
ax5 = fig.add_subplot(2, 3, 5)
ax6 = fig.add_subplot(2, 3, 6)
ax1.imshow(H, cmap='gray')
ax1.set_title("H")
ax1.axis('off')
ax2.imshow(S, cmap='gray')
ax2.set_title("S")
ax2.axis('off')
ax3.imshow(V, cmap='gray')
ax3.set_title("V")
ax3.axis('off')
ax4.imshow(R_image)
ax4.set_title("R")
ax4.axis('off')
ax5.imshow(G_image)
ax5.set_title("G")
ax5.axis('off')
ax6.imshow(B_image)
ax6.set_title("B")
ax6.axis('off')
plt.show()
```
