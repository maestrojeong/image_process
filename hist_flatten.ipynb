```python
import matplotlib.pyplot as plt
from ops import rgb2gray
from ops import flatten_img
```

```python
image = plt.imread('./image.png')
gray_image = rgb2gray(image, option = 0)
```

```python
hist, cum_hist, norm_cum_hist, hist2, flat_image = flatten_img(gray_image, option = True)
```

```python
fig = plt.figure(figsize=(17,17))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.plot(hist, 'r-')
ax1.set_title("histogram")
ax2.plot(cum_hist, 'g-')
ax2.set_title("cumul_histogram")
ax3.plot(norm_cum_hist, 'b-')
ax3.set_title("Normed_cum_histogram")
ax4.plot(hist2, 'k-')
ax4.set_title("flat_histogram")
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(17,17))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.imshow(gray_image, cmap='gray')
ax1.set_axis_off()
ax2.imshow(flat_image, cmap='gray')
ax2.set_axis_off()
plt.show()
```
