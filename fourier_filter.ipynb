```python
import matplotlib.pyplot as plt
import numpy as np
from ops import rgb2gray
from ops import freq_filter
from ops import clip_2d
```

```python
image = plt.imread('./image.png')
gr_img = rgb2gray(image)
plt.imshow(gr_img, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
low_pass, high_pass = freq_filter(gr_img, 50.0)
```

```python
fig = plt.figure(figsize=(17,17))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.imshow(low_pass, cmap='gray')
ax1.set_axis_off()
ax1.set_title("Low pass filter")
ax2.imshow(clip_2d(high_pass-np.min(high_pass)), cmap='gray')
ax2.set_axis_off()
ax2.set_title("High pass filter")
plt.show()
```
