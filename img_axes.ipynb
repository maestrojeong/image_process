```python
import matplotlib.pyplot as plt
import numpy as np
from ops import get_HSI
from ops import get_HSV
from ops import get_HSL
```

```python
image = plt.imread('./dark.png')
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
H, S, V = get_HSV(image)
H1, S1, L1 = get_HSL(image)
H2, S2, I2 = get_HSI(image)
```

```python
height, width, channel = image.shape
R_image = np.zeros((height,width, channel))
G_image = np.zeros((height,width, channel))
B_image = np.zeros((height,width, channel))
for h in range(height):
    for w in range(width):
        R_image[h][w][0] = image[h][w][0]
        G_image[h][w][1] = image[h][w][1]
        B_image[h][w][2] = image[h][w][2]
fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)
ax1.imshow(R_image)
ax1.set_title("R")
ax1.axis('off')
ax2.imshow(G_image)
ax2.set_title("G")
ax2.axis('off')
ax3.imshow(B_image)
ax3.set_title("B")
ax3.axis('off')
plt.show()
```
